```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load your dataset (replace 'your_file.csv' with the actual file name after uploading)
data = pd.read_csv('/content/sample_data/Mobile_Price_Classification-220531-204702.csv')

# Features (all columns except 'price_range')
X = data.drop('price_range', axis=1)

# Labels (target: 'price_range')
y = data['price_range']

# Split the data into training and testing sets (75% train, 25% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Normalize the data (standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build the ANN model
model = Sequential()

# Adding the first hidden layer with 8 neurons and ReLU activation function
model.add(Dense(units=8, activation='relu', input_dim=X_train.shape[1]))

# Adding the second hidden layer with 4 neurons and ReLU activation function
model.add(Dense(units=4, activation='relu'))

# Adding the output layer with 1 neuron and Sigmoid activation function (binary classification)
model.add(Dense(units=1, activation='sigmoid'))

# Compile the ANN
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the ANN model with 100 epochs and batch size of 32
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Save the model weights
model.save_weights('ann_weights.weights.h5') # Changed the filename to include '.weights.h5'

# Print the model summary
print(model.summary())

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_acc}')
```
